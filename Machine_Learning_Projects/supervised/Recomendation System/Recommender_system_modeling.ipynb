```python
def convert(gen):
  list =[]
  for i in ast.literal_eval(gen):
    list.append(i['name'])
  return list
```

I have an issue this list all are "string"

TypeError: string indices **must be integers**
im using __ast_literal_eval__ function

```python
df['genres'] = df['genres'].apply(convert)
```

```python
df.head(3)
```

```python
df['keywords'] = df['keywords'].apply(convert)
```

```python
df.head(3)
```

```python
def convert2(gen):
  list =[]
  counter = 0
  for i in ast.literal_eval(gen):
    if counter !=3:
      list.append(i['name'])
      counter+=1
    else:
      break
  return list
```

```python
df['cast'] = df['cast'].apply(convert2)
```

```python
def director(gen):
  list =[]
  for i in ast.literal_eval(gen):
    if i['job'] == 'Director':
      list.append(i['name'])
      break
  return list
```

```python
df['crew'] = df['crew'].apply(director)
```

```python
df['overview'] = df['overview'].apply(lambda x: x.split())
```

```python
df['genres'] = df['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
df['keywords'] = df['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
df['cast'] = df['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
df['crew'] = df['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
```

```python
new_df = df[['movie_id', 'title', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

new_df['tags'] column each words with dimention like XY graph 5000 vector lines are their. so each close vectors will recomended.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

in the array have tags (words) and some numerical, alfa numerical terms are there.

plural, present, past, future words are there will take nuteral and combined common words the process called **steming**.

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
  j=[]
  for i in text.split():
    j.append(ps.stem(i))
  return " ".join(j)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

similarity have Index issue on single movie.
for example '0' index movie similar movies have no indexs.
im using enumerate function here

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

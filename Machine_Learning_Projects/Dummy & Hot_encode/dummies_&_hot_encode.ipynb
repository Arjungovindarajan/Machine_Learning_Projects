```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
dumm = pd.get_dummies(df.town)
dumm
```

```python
merged = pd.concat([df,dumm], axis = 'columns')
merged
```

```python
out = merged.drop(['town','west windsor'], axis = 'columns')
out
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
```

```python
X = out.drop('price',axis ='columns')
```

```python
y = out.price
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.score(X,y)
```
